```python
from pytube import Channel
import pandas as pd
import re

# Define the function to extract video ID from the Channel URL
def extract_video_id(channel_url):
    try:
        # Create a Channel object
        channel = Channel(channel_url)

        # Access the HTML content
        html_content = channel.html

        # Use regular expression to extract video ID
        result = re.search(r'videoRenderer":{"videoId":(.*?),', html_content)
        
        # Return the extracted video ID or None if not found
        return result.group(1) if result else None
    except Exception as e:
        print(f"Error processing channel URL: {channel_url}. Error: {e}")
        return None

# Apply the function to create a new column 'V_id'
df['V_id'] = df['Channel_url'].apply(extract_video_id)

df
```

```python
# Create a new column 'V_url' by concatenating the base YouTube URL with 'V_id'
df['V_url'] = 'https://www.youtube.com/watch?v=' + df['V_id']

# Remove quotes from 'V_url' column
df['V_url'] = df['V_url'].str.replace('"', '')

df
```

```python
from pytube import YouTube
import pandas as pd
from datetime import datetime

def get_publish_date(V_url):
    try:
        yt = YouTube(V_url)
        publish_date = yt.publish_date
        return publish_date
    except Exception as e:
        print(f"Error getting publish date for {V_url}. Error: {e}")
        return None

# Apply the function to create a new column 'Publish_Date'
df['Publish_Date'] = df['V_url'].apply(get_publish_date)

# Convert the 'Publish_Date' column to only date format
df['Publish_Date'] = pd.to_datetime(df['Publish_Date']).dt.date

df
```
