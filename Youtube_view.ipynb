```python
df.dropna(inplace=True)
```

```python
# Extract numeric part using string manipulation
df['Subscriber'] = df['Subscriber'].str.split(' ').str[0]
df
```

```python
# Function to convert subscriber counts to numeric values
def convert_to_numeric(subscriber_count):
    if 'K' in subscriber_count:
        return float(subscriber_count.replace('K', '')) * 1000
    elif 'M' in subscriber_count:
        return float(subscriber_count.replace('M', '')) * 1000000
    else:
        return float(subscriber_count)

# Apply the conversion function to the 'Subscribers' column
df['Subscriber'] = df['Subscriber'].apply(convert_to_numeric)

df
```

```python
# Sort the DataFrame in decreasing order based on 'Subscribers_Num'
df_sorted = df.sort_values(by='Subscriber', ascending=False)

df_sorted.head(30)
```

```python
# Replace multiple patterns in the 'Channel_url' column
df_sorted['Channel_url'] = df_sorted['Channel_url'].str.replace('@', 'c/')

df_sorted
```

```python
# Assuming df_sorted has a 'Channel_id' column
df_sorted.drop_duplicates(subset='Channel_url', keep='first', inplace=True)

df_sorted
```

```python
# Assuming df_sorted is your DataFrame
df_sorted.drop(df_sorted.index[100:], inplace=True)

df_sorted
```

```python
from pytube import YouTube
from pytube import Channel
import pandas as pd
from urllib.error import HTTPError

# Define the get_channel_id function with error handling
def get_channel_id(channel_url):
    try:
        # Create a Channel object
        channel = Channel(channel_url)
        print(channel_url)

        # Access the channel ID
        channel_id = channel.channel_id

        return channel_id
    except HTTPError as e:
        print(f"Error accessing channel URL: {channel_url}. HTTP Error: {e}")
        return None


# Apply the get_channel_id function to the 'Channel_url' column
df_sorted['Channel_id'] = df_sorted['Channel_url'].apply(get_channel_id)

# Display the modified DataFrame
df_sorted
```

```python
df_sorted.dropna(inplace=True)
```

```python
# Assuming df_sorted has a 'Channel_id' column
df_sorted.drop_duplicates(subset='Channel_id', keep='first', inplace=True)

df_sorted
```
